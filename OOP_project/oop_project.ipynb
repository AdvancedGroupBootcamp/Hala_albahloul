```python
import pandas as pd
from datetime import date
class Computer():
    
    def __init__(self,manufacturer=None,model=None,category=None,screenSize=None,screen=None,CPU=None,RAM=None, storage=None,
                 GPU=None,operatingSystem=None,operatingSystemVersion=None,weight=None,price=None,remaining=0 ):
        self.manufacturer=manufacturer
        self.model=model
        self.category=category
        self.screenSize=screenSize
        self.screen=screen
        self.CPU=CPU
        self.frequency=None
        self.RAM=RAM
        self.storage=storage
        self.GPU=GPU
        self.operatingSystem=operatingSystem
        self.operatingSystemVersion=operatingSystemVersion
        self.weight=weight
        self.price=price
        self.remaining=remaining
        self.purchases=0
        
    def get_price(x):
        return x.price
    
    def get_model(self):
        return self.model
    
    def set_new_price(self, new_price):
        self.price=new_price
        
        
    def data_featuring(self):
        self.frequency=self.CPU.split(" ")[3]
        
    def data_extracting(self,form):
        self.manufacturer=form.split(",")[0]
        self.model=form.split(",")[1]
        self.category=form.split(",")[2]
        self.screenSize=form.split(",")[3]
        self.screen=form.split(",")[4]
        self.CPU=form.split(",")[5]
        self.RAM=form.split(",")[6]
        self.storage=form.split(",")[7]
        self.GPU=form.split(",")[8]
        self.operatingSystem=form.split(",")[9]
        self.operatingSystemVersion=form.split(",")[10]
        self.weight=form.split(",")[11]
        self.price=form.split(",")[12]
        self.remaining=int(form.split(",")[13])
        
        
    def get_data_cleaned(self):
        return{
        "manufacturer":self.manufacturer,
        "model":self.model,
        "category":self.category,
        "screenSize":self.screenSize,
        "screen":self.screen,
        "CPU":self.CPU,
        "frequency":self.frequency,
        "RAM":self.RAM,
        "storage":self.storage,
        "GPU":self.GPU,
        "operatingSystem":self.operatingSystem,
        "operatingSystemVersion":self.operatingSystemVersion,
        "weight":self.weight,
        "price":self.price,
        "remaining":self.remaining,
        "purchases":self.purchases
            
        }
    
        
```
